# Start spark session

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("bank").getOrCreate()
```

# Read the dataset

```python
import pandas as pd
pd_df = pd.read_excel('/content/18. Sample Data-2.xlsx')
df = spark.createDataFrame(pd_df)
```

```python
df.show()
```

# Check the datatype of the dataframe

```python
df.printSchema()
```

# Check the Summary

```python
df.describe()
```

# Select Columns

```python
df.select(['Customer Name', 'Category', 'Sales']).show()
```

# Adding Columns

```python
df_revised = df.withColumn('sales_after_discount', df['Sales'] - df['Discount'])
```

```python
df_revised.show()
```

# Dropping Columns

```python
df_final = df_revised.drop('age_after_10_years')
```

```python
df_final.show()
```

# Renaming Columns

```python
df_final = df_final.withColumnRenamed('job','profession')
```

```python
df_final.show()
```
