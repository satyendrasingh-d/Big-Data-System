# Logical Operators

```python
from pyspark.sql.functions import col
df.filter(~(col('Retailer ID') > 1185732) & (col('Retailer ID') < 1200000)).show()
```

```python
from pyspark.sql.functions import col
df.filter((col('Retailer ID') > 1185732) | (col('Retailer ID') < 1185732)).show()
```

# Groupby

```python
from pyspark.sql.functions import col, regexp_replace, sum

# Clean 'Units Sold' by removing commas and cast to integer
units_sold_cleaned = regexp_replace(col('Units Sold'), ',', '').cast('int')

# Group by 'Region' and sum the cleaned 'Units Sold'
df.groupBy('Region').agg(sum(units_sold_cleaned).alias('Total_Units_Sold')).show()
```

# |Overall_Total_Units_Sold

```python
from pyspark.sql.functions import col, regexp_replace, sum

# Clean 'Units Sold' by removing commas and cast to integer
units_sold_cleaned = regexp_replace(col('Units Sold'), ',', '').cast('int')

# Calculate the overall sum of cleaned 'Units Sold'
overall_total_units_sold = df.agg(sum(units_sold_cleaned).alias('Overall_Total_Units_Sold'))

overall_total_units_sold.show()
```

# Format_number, Alias

```python
from pyspark.sql.functions import col, regexp_replace, format_number

# Clean 'Total Sales' by removing '$' and ',' and cast to double
cleaned_total_sales = regexp_replace(col('Total Sales'), '[$,]', '').cast('double')

# Apply format_number to the cleaned total sales
df.withColumn("Formatted Total Sales", format_number(cleaned_total_sales, 2)).show()
```

```python
from pyspark.sql.functions import col

# Use alias to rename 'Region' to 'Sales Region' and 'Retailer ID' to 'Retailer_Identifier'
df.select(
    col('Region').alias('Sales_Region'),
    col('Retailer ID').alias('Retailer_Identifier'),
    col('Product')
).show()
```

# Sort data using orderBy

```python
from pyspark.sql.functions import col, regexp_replace

# Clean 'Total Sales' by removing '$' and ',' and cast to double
cleaned_total_sales_for_sort = regexp_replace(col('Total Sales'), '[$,]', '').cast('double')

# Sort by 'Total Sales' in ascending order
print("DataFrame sorted by 'Total Sales' (Ascending):")
df.orderBy(cleaned_total_sales_for_sort.asc()).show()
```

```python
from pyspark.sql.functions import col, regexp_replace

# Clean 'Units Sold' by removing commas and cast to integer
cleaned_units_sold_for_sort = regexp_replace(col('Units Sold'), ',', '').cast('int')

# Sort by 'Units Sold' in descending order
print("DataFrame sorted by 'Units Sold' (Descending):")
df.orderBy(cleaned_units_sold_for_sort.desc()).show()
```
