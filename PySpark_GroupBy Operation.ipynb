# GroupBy Operation

# Group by department and count employees in each department.

```python
# Group by department and count employees in each department
dept_count = df.groupBy("Department").count()
dept_count.show()
```

# Find total salary per department.

```python
salary_dept = df.groupBy("Department").sum("Salary")
salary_dept.show()
```

# Find average salary per department.

```python
avg_salary = df.groupBy('Department').avg('Salary')
avg_salary.show()
```

# Group by city and find max salary.

```python
from pyspark.sql.functions import max
max_salary = df.groupBy("city").agg(max("salary").alias("highest_salary"))
max_salary.show()
```

# Group by department and gender and count employees.

```python
dept_gender_count = df.groupBy("Department", "Gender").count()
dept_gender_count.sort("Department").show()
```

# Multiple Aggregation

# Perform multiple aggregations on salary per department:

```python
from pyspark.sql.functions import max, min, sum, avg, max, min, countDistinct, desc

dept = df.groupBy("Department").agg(
    sum("Salary").alias("Total_salary"),
    avg("Salary").alias("Average_salary"),
    max("Salary").alias("Max_salary"),
    min("Salary").alias("Min_salary")
)

dept.show()
```

# Rename all aggregated columns appropriately.

```python
from pyspark.sql.functions import sum, avg, max, min, count, desc

# Performing multiple aggregations and renaming columns using .alias()
dept_salary_stats = df.groupBy("department").agg(
    sum("salary").alias("Total_Dept_Salary"),
    avg("salary").alias("Average_Dept_Salary"),
    max("salary").alias("Max_Dept_Salary"),
    min("salary").alias("Min_Dept_Salary"),
    count("Emp_id").alias("Employee_Count")
)

# Display the results
dept_salary_stats.show()
```

# Group by department and role, then calculate:
* total salary
* average salary
* employee count


```python
from pyspark.sql.functions import sum, avg, count

role = df.groupBy("department", "role").agg(
    sum("salary").alias("total_role_salary"),
    avg("salary").alias("avg_role_salary"),
    count("emp_id").alias("emp_count")
)

role.show()
```

# Count distinct roles in each department.

```python
# Count distinct roles per department
distinct_roles_df = df.groupBy("department").agg(
    countDistinct("role").alias("unique_roles")
)
distinct_roles_df.show()
```

# Find the top 3 departments with the highest total salary.

```python
# Find the Top 3 Departments by total salary expenditure
top_3_depts = df.groupBy("department") \
    .agg(sum("salary").alias("total_spend")) \
    .orderBy(desc("total_spend")) \
    .limit(3)

top_3_depts.show()
```
